```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from wordcloud import WordCloud
```

```python
data = pd.read_csv('GMaps_Review_Classified.csv')
```

```python
data.info()
```

```python
data['Nama Tempat'].value_counts()
```

```python
import re

# daftar
stopwords = ['untuk', 'yang', 'dan', 'juga','nya', 'di', 'yg', 'tapi', 'ini','dengan','karena','atau','ada','jadi','banget',
             'bisa','kalau','dari','jika','sama','gak','kalo','sih','aja','sih','mau','buat','selalu','banyak','saya','tidak','bagi']

# function
def remove_stopwords(text):
    if isinstance(text, str):  # make sure string
        pattern = r'\b(?:' + '|'.join(stopwords) + r')\b'
        return re.sub(pattern, '', text, flags=re.IGNORECASE).strip()
    return text


data['Review'] = data['Review'].apply(remove_stopwords)
```

## Umum

```python
plt.figure(figsize=(10, 6))
# Drop NaN values from the 'Review' column
reviews = data['Review'].dropna()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(reviews))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()
```

## Goa Gong

```python
tempat = 'Goa Gong Pacitan'  
filtered_data = data[data['Nama Tempat'] == tempat]

filtered_reviews = filtered_data['Review'].dropna()
text = ' '.join(filtered_reviews)

# Buat WordCloud
plt.figure(figsize=(10, 6))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title(f'Word Cloud for {tempat}')
plt.show()
```
