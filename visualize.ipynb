```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('GMaps_Review_Classified.csv')
data.head()
```

```python
# bar chart distribusi labels
plt.figure(figsize=(10, 6))
sns.countplot(x='labels', data=data)
plt.title('Distribusi Emosi')
plt.xlabel('Emosi')
plt.ylabel('Jumlah')
plt.show()
```

```python
data['labels'].value_counts()
```

```python
# rincian jumlah data per tempat
data['Nama Tempat'].value_counts()
```

```python
# stack bar chart untuk melihat distribusi label berdasar wistawan lokal
plt.figure(figsize=(10, 6))
sns.countplot(x='labels', hue='Wisatawan Lokal', data=data)
plt.title('Distribusi Emosi Berdasarkan Wisatawan Lokal')
plt.xlabel('Emosi')
plt.ylabel('Jumlah')
plt.legend(title='Wisatawan Lokal')
plt.show()
```

```python
data['labels'].count()
```

```python
# bar chart distribusi di setiap tempat
for place in data['Nama Tempat'].unique():
    sns.countplot(x='labels', hue='Wisatawan Lokal',data=data[data['Nama Tempat'] == place])
    plt.title(f'Distribusi Emosi di {place}')
    plt.xlabel('Emosi')
    plt.ylabel('Jumlah')
    plt.tight_layout()
    plt.show(block=False)
    plt.clf()
```

## Goa Gong Pacitan

```python
# filter nama tempat
nama_tempat = 'Goa Gong Pacitan'
filtered_data = data[data['Nama Tempat'] == nama_tempat]

kategori_counts = filtered_data['labels'].value_counts()

# Membuat pie chart
plt.figure(figsize=(8, 8))  # Ukuran pie chart
wedges, texts, autotexts = plt.pie(
    kategori_counts, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=plt.cm.Paired.colors,
    textprops={'fontsize': 8}  #  font untuk persentase
)
plt.legend(wedges, kategori_counts.index, title="Kategori", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title(f'Distribusi Kategori Emosi Wisatawan di {nama_tempat}')  
plt.axis('equal') # lingkaran bulat
plt.show()
```
